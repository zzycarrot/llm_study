```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
words =  open('names.txt','r').read().splitlines()
words[:10]
```

```python
N = torch.zeros((27,27),dtype = torch.int32)
chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}
itos
```

```python
#build dataset
block_size = 3
X, Y = [], []
for word in words:
    # print(word)
    context = [0] * block_size
    for ch in word + '.':
        ix = stoi[ch]
        X.append(context)
        Y.append(ix)
        # print(''.join(itos[i] for i in context),'--->',itos[ix])
        context = context[1:] + [ix]

X = torch.tensor(X)
Y = torch.tensor(Y)
```

```python
X.shape
```

```python
#lookup table | embbeding
C = torch.randn((27, 2))
```

```python
W1 = torch.randn((3 * 2, 100))
b1 = torch.randn(100)
```

```python
W2 = torch.randn((100, 27))
b2 = torch.randn(27)
```

```python
param = [C, W1, b1, W2, b2]
```

```python
epoch = 10000
lre = torch.linspace(-3,0,epoch)
lrs = 10**lre
```

```python
for p in param:
        p.requires_grad = True
lossi, lri = [], []  
for i in range(epoch):
    #minibatch
    ix = torch.randint(0, X.shape[0], (32,))
    #forward pass
    emb = C[X[ix]]
    h = torch.tanh(emb.view(emb.shape[0], 6) @ W1 + b1)
    logits = h @ W2 + b2
    loss = F.cross_entropy(logits, Y[ix])
    #backward
    for p in param:
        p.grad = None
    loss.backward()
    #updata
    lr = 0.1
    for p in param:
        p.data += -lr * p.grad
    # lri.append(lre[i])
    # lossi.append(loss.item())
loss = F.cross_entropy(logits, Y[ix])
loss.item()
```

```python
emb = C[X]
h = torch.tanh(emb.view(emb.shape[0], 6) @ W1 + b1)
logits = h @ W2 + b2
loss = F.cross_entropy(logits, Y)
loss.item()
```

```python
g = torch.Generator().manual_seed(2147483647 + 10)

for _ in range(20):
    out = []
    context = [0] * block_size #  '...'
    
    while True:
        # ---  ---
        # 1.  context  Embedding
        emb = C[torch.tensor([context])] #  (1, block_size, embedding_dim)
        
        # 2.  ( view )
        #  6  block_size(3) * embedding_dim(2) = 6
        h = torch.tanh(emb.view(1, -1) @ W1 + b1)
        
        # 3.  logits
        logits = h @ W2 + b2
        
        # 4. Softmax 
        probs = F.softmax(logits, dim=1)
        
        # ---  ---
        ix = torch.multinomial(probs, num_samples=1, generator=g).item()
        
        # ---  context  ---
        context = context[1:] + [ix]
        
        if ix == 0: #  '.' 
            break
        out.append(itos[ix])
        
    print(''.join(out))
```
